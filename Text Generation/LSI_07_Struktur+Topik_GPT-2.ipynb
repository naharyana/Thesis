# Seed Text

```python
# load doc into memory
def load_doc(filename):
    # open the file as read only
    file = open(filename, 'r')
    # read all text
    text = file.read()
    # close the file
    file.close()
    return text
```

```python
from random import randint
# load cleaned text sequences
in_filename = '/content/drive/My Drive/data/spam23.txt'
doc = load_doc(in_filename)
lines = doc.split('\n')
seq_length = len(lines[0].split()) - 1
```

```python
len(lines)
```

```python
#untuk gpt2

seed=[]

for i in lines:
    # select a seed text
    seed_text = i #randint(0,len(i)/2)
    text = seed_text.split()
    n = int(len(text)/2)+1
    text_split=[' '.join(text[i:i+n]) for i in range(0,len(text),n)]
    seed.append(text_split)
    print(text_split[0] + '\n')
```

```python
prompts = [i[0] for i in seed]
```

```python
for i in prompts:
  print(i)
```

# Generation

```python
gen=[]
import logging
from simpletransformers.language_generation import LanguageGenerationModel


logging.basicConfig(level=logging.INFO)
transformers_logger = logging.getLogger("transformers")
transformers_logger.setLevel(logging.WARNING)

gen_args = {
    "do_sample": True,
    "block_size": 512,
    "stop_token": None,
}

model = LanguageGenerationModel("gpt2", "/content/drive/My Drive/GPT-2/model/gpt2_d0d1_LSI2", args=gen_args, use_cuda=False)

for prompt in prompts:
    # Generate text using the model. Verbose set to False to prevent logging generated sequences.
    generated = model.generate(prompt, verbose=False)
    #print(generated)
    generated = '.'.join(generated[0].split('.')[:-1]) + '.'
    if (generated == "") or (generated == prompt):
      generated = model.generate(prompt, verbose=False)
      generated = '.'.join(generated[0].split('.')[:-1]) + '.'
    gen.append(generated)
```
